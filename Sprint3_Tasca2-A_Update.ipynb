<span style=color:blue>*Selección de canales R, G, B
https://stackoverflow.com/questions/37431599/how-to-slice-an-image-into-red-green-and-blue-channels-with-misc-imread*.</span>

```python
# eliminar un color(R,G o B) de la imagen...Red=0, Green=1, Blue=2
ocell_woR = ocell.copy() # Duplicate image
ocell_woR[:, :, (0)] = 0    # Zero out contribution from red
plt.imshow(ocell_woR)
```

```python
mpimg.imsave('ocell_sense_vermell.jpeg',ocell_woR)
```

```python
# generar imagen monocolor eliminandodos de los 3 colores....without R and G
ocell_woRandG = ocell.copy() # Duplicate image
ocell_woRandG[:, :, (0,1)] = 0    # Zero out contribution from red & green
plt.imshow(ocell_woRandG)
```

```python
# generar imagen monocolor eliminandodos de los 3 colores....without G and B
ocell_wogRandB = ocell.copy() # Duplicate image
ocell_wogRandB[:, :, (1,2)] = 0    # Zero out contribution from green and Blue
plt.imshow(ocell_wogRandB)
```

```python
mpimg.imsave('ocell_sense_verd-blau.jpeg',ocell_wogRandB)
```

```python
# Guardar
titles = ['Ocell', 'Red channel', 'Green channel', 'Blue channel']
cmaps = [None, plt.cm.Reds_r, plt.cm.Greens_r, plt.cm.Blues_r]

fig, axes = plt.subplots(1, 4, figsize=(13,4))
objs = zip(axes, (ocell, *ocell.transpose(2,0,1)), titles, cmaps)

for ax, channel, title, cmap in objs:
    ax.imshow(channel, cmap=cmap)
    ax.set_title(title)
    ax.set_xticks(())
    ax.set_yticks(())

plt.savefig('RGB1.png')
```
